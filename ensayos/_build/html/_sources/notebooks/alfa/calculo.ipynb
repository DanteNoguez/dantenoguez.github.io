````python
import numpy as np
import matplotlib.pyplot as plt
````

Si bien la pendiente calcula la inclinación de una recta, lo conceptualmente importante de la inclinación es que podemos interpretarla como la proporción del cambio que hay entre $x$ y $y$. En otras palabras, qué tanto cambia $y$ cuando cambiamos $x$. Supongamos, por ejemplo, la recta de una función que podemos definir como $f(x) = y = xb$, donde:

````python
x = np.arange(0, 21, 2) 
b = 5
def f(x, b): return x*b
c = f(x, b)

fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(x, c, marker='o', color='blue') 
plt.xticks(ticks=x)
ax.set_facecolor('black')
ax.set_xlabel('Valores de x')
ax.set_ylabel('Valores de c')
ax.hlines(y=10, xmin=0, xmax=2, linewidth=1, color='white', linestyles='dashed')
ax.vlines(x=2, ymin=0, ymax=10, linewidth=1, color='white', linestyles='dashed')
plt.show()
````

Ahora, para calcular la pendiente según la fórmula

$$
Pendiente = \frac{y_2 - y_1}{x_2 - x_1} = \frac{\Delta y}{\Delta x}
$$

```{margin}
Los símbolos $\Delta, \delta$ son letras del alfabeta griego llamadas «delta», y generalmente se utilizan en matemáticas para representar o leer un cambio: «cambio en $y$ sobre cambio en $x$».
```

Únicamente debemos tomar dos puntos cualesquiera y reemplazar nuestros valores. Por ejemplo, si consideramos los valores del punto denotado por las líneas punteadas, de tal manera que $x_1 = 2, y_1 = 10$, podemos tomar el punto siguiente para calcular la inclinación, de manera que

$$
Pendiente = \frac{20 - 10}{4 - 2} = \frac{10}{2} = 5
$$

Pero las funciones lineales —es decir, las rectas—, por definición tienen la misma inclinación en todos sus puntos o segmentos. Comprobémoslo cambiando nuestras $x_2, y_2$:

$$
Pendiente = \frac{60 - 10}{12 - 2} = \frac{50}{10} = 5
$$

La pendiente, es decir, la inclinación de la recta es la misma. Ahora, ¿exactamente qué quiere decir el número cinco? Que por cada paso que damos en $x$, avanzamos cinco veces más $y$ o, dicho de otra forma, que $y$ es el resultado de multiplicar $x$ por cinco, o que $x$ influye con una «fuerza» de cinco veces sí mismo en $y$.

El problema fundamental de la pendiente es que solo es válida para una recta o función lineal, puesto que una función no lineal no satisface la propiedad de tener la misma inclinación en todos sus puntos, sino todo lo contrario. Primero, ejemplifiquemos con la pendiente de una función cuadrática.

````python
pendiente2 = (c[6] - c[3]) / (x[6] - x[3])

print(f'Valor de la pendiente 2: {pendiente2}')
fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(x, c, marker='o', color='blue', label='recta de la función') 
plt.xticks(ticks=x)
ax.set_facecolor('black')
ax.set_xlabel('Valores de x')
ax.set_ylabel('Valores de c') 
ax.axline((x[6], c[6]), slope=pendiente2, color='green', label='pendiente 2') 
ax.hlines(y=10, xmin=0, xmax=2, linewidth=1, color='white', linestyles='dashed')
ax.vlines(x=2, ymin=0, ymax=10, linewidth=1, color='white', linestyles='dashed')
ax.legend()
plt.show()
````

Sin embargo, de aquí podemos deducir algo: mientras la distancia de los valores de $x$ sea menor, nuestra pendiente es más acertada. Dicho de otra forma, 

````python
def cuadratica(x): return x**2

c = cuadratica(x)

pendiente1 = (c[6] - c[5]) / (x[6] - x[5])
fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(x, c, marker='o', color='blue', label='recta de la función') 
plt.xticks(ticks=x)
ax.set_facecolor('black')
ax.set_xlabel('Valores de x')
ax.set_ylabel('Valores de f(x)')
ax.axline((x[6], c[6]), slope=pendiente1, color='red', label='pendiente 1') 
ax.hlines(y=c[6], xmin=0, xmax=x[6], linewidth=1, color='white', linestyles='dashed')
ax.hlines(y=c[5], xmin=0, xmax=x[5], linewidth=1, color='white', linestyles='dashed')
ax.vlines(x=x[6], ymin=0, ymax=c[6], linewidth=1, color='white', linestyles='dashed')
ax.vlines(x=x[5], ymin=0, ymax=c[5], linewidth=1, color='white', linestyles='dashed')
ax.legend()
plt.show()
````

```{figure} ../../img/derivada.png
La derivada se representa mediante la tangente porque es equivalente a una secante con una distancia infinitamente pequeña entre dos puntos, tan pequeña que es similar a un solo punto. Esa secante pequeñísima tendría una inclinación equivalente a la de la tangente.
```

En esencia, quiero minimizar mis distancias entre $x$ e $y$ para tener un resultado más riguroso. Suponiendo que llamemos $h$ a la distancia entre $x_2$ y $x_1$, nuestra fórmula actual para esta pendiente luce así:

$$
Pendiente = \frac{\Delta y}{\Delta x} = \frac{f(x+h) - f(x)}{(x+h) - x} = \frac{f(x+h) - f(x)}{h}
$$

Pero, idealmente, queremos que $h$ sea lo más pequeño posible, es decir, lo más cercano a 0 posible —sin ser 0, puesto que queremos dos $x$ tales que nos permitan determinar la pendiente—. Para expresar esta idea, emplearemos la expresión $\lim _{h \rightarrow 0}$, es decir, «cuando $h$ tiende a 0» o, más estrictamente, «el límite de la función cuando $h$ tiende a 0». De tal manera que:

$$
Pendiente = \lim _{h \rightarrow 0} \frac{f(x+h) - f(x)}{h}
$$

Pero ahora ya no hablamos de una pendiente exactamente, sino que la estamos alterando para conseguir un resultado distinto. A esta nueva fórmula podemos denominarla derivada:

$$
Derivada = \frac{dy}{dx} = f'(x) = \lim _{h \rightarrow 0} \frac{f(x+h) - f(x)}{h}
$$

Y de la misma forma que la derivada es la pendiente, podemos entender a la integral definida como el área debajo de un segmento de la función. Por ejemplo:

````python
x = np.arange(0, 21, 2) 
b = 5
def f(x, b): return x*b
c = f(x, b)

fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(x, c, marker='o', color='blue') 
plt.xticks(ticks=x)
ax.set_facecolor('black')
ax.set_xlabel('Valores de x')
ax.set_ylabel('Valores de c')
ax.hlines(y=c[3], xmin=x[3], xmax=x[6], linewidth=1, color='white', linestyles='dashed')
ax.vlines(x=x[3], ymin=0, ymax=c[3], linewidth=1, color='white', linestyles='dashed')
ax.vlines(x=x[6], ymin=0, ymax=c[6], linewidth=1, color='white', linestyles='dashed')
plt.show()
````

La integral determinaría el área desde $x=6$ hasta $x=12$, y para abordar este problema geométricamente podemos dividir al área en dos figuras tales que:

$$
Rectángulo = base \times altura = (12 - 6)(30) = 180
$$

$$
Triángulo = \frac{1}{2} \times base \times altura = \frac{base \times altura}{2} = \frac{(12-6)(60-30)}{2} = 90
$$

$$
\text{Área total} = 180 + 90 = 270
$$

Pero podemos plantear el problema también así:

$$
\int_{a}^{b}f(x)dx 
$$

Y para resolverlo, utilizar la fórmula:

$$
\int_{a}^{b}x^ndx = \frac{x^{n+1}}{n + 1} + C, \ n \neq -1
$$

Utilizando una constante $C = 0$, tenemos que:

$$
\int_{x=6}^{x=12}(x^1 \cdot b)dx = \frac{x^2 \cdot 5}{2} = \frac{12^2 \cdot 5}{2} - \frac{6^2 \cdot 5}{2} = \frac{144 \cdot 5}{2} - \frac{36 \cdot 5}{2} = \frac{720}{2} - \frac{180}{2} = 360 - 90 = 270
$$

La diferencia sustancial es la misma que con las derivadas: obtener el área debajo de un segmento de una función no lineal mediante figuras geométricas sería inexacto y tedioso, de manera que el cálculo integral resuelve este problema.

### Teorema fundamental del cálculo

Por último, el teorema fundamental del cálculo estatuye que la derivación y la integración son operaciones inversas. Mientras que la derivación consiste fundamentalmente en una operación de división y con ella obtenemos —en dos dimensiones— la pendiente de la tangente a un punto, la integración fundamentalmente consiste en una multiplicación para obtener el área bajo la curva de la función entre dos puntos dados.

### Reflexiones

Curiosamente, la palabra «cálculo» viene del latín [*calculus*,](https://www.perseus.tufts.edu/hopper/text?doc=Perseus:text:1999.04.0059:entry=calculus) cuyo significado es «pequeña piedra». Calcular, en las épocas antiguas, implicaba el uso de piedritas que se iban contando una tras otra, justo como cuando éramos pequeños hacíamos con el ábaco.

El cálculo lidia con cambios infinitamente pequeños —de ahí su nombre «cálculo infinitesimal»—. Inicialmente, estos «cambios pequeños» se concibieron como trucos prácticos que facilitaban cálculos como los que hemos hecho, pero nunca hubo una justificación matemáticamente rigurosa para proceder así, «imaginando» secantes infinitamente pequeñas. De ahí precisamente que Newton se hubiera negado a publicar su trabajo de cálculo y que Leibniz tiempo después afirmara que lidiar con lo infinitamente pequeño era una simple ficción mental que facilitaba ciertos cálculos en la práctica.
